# Distiller Android APK 构建
1. 点击上方「代码执行程序」→「全部运行」
2. 等待约 30 分钟
3. 完成后左侧文件列表会出现 distiller-*.apk
4. 右键下载到电脑

```
# 安装依赖
!sudo apt update -qq
!sudo apt install -y -qq git zip unzip openjdk-17-jdk python3-pip autoconf libtool pkg-config zlib1g-dev libncurses5-dev libncursesw5-dev cmake libffi-dev libssl-dev libltdl-dev
!pip install buildozer cython -q
```

```
# 克隆项目
!rm -rf distiller
!git clone https://github.com/shendiaox/distiller.git
%cd distiller
!cp android_app.py main.py
```

```
# 构建 APK（约 25-40 分钟）
!buildozer android debug
```

```
# 列出生成的 APK
import glob
apks = glob.glob('**/*.apk', recursive=True) + glob.glob('bin/*.apk')
for a in apks:
    import os
    size_mb = os.path.getsize(a) / 1024 / 1024
    print(f'APK: {a} ({size_mb:.1f} MB)')
apks
```
